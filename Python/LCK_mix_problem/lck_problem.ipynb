```python
import numpy as np
import pandas as pd

#read dataset
df_players = pd.read_csv("lck_players.csv")

print("Team list :")
print(set(df_players["Team"]))

print("Players")
for team in set(df_players["Team"]):
    print("\n", team)
    print(df_players[df_players["Team"] == team])
```

```python
from itertools import *

teams = list(set(df_players["Team"]))

total_case = 0

for team in teams:
    each_team = df_players[df_players["Team"] == team]
    each_team_position = list(each_team["Name"])
    each_team_case = 0

    print(team, "Position [TOP, JUNGLE, MID, AD, SUPPORT]")
    print(each_team_position, "\n")

    case_list = list(permutations(each_team["Name"], 5))
    
    print("Cases when every player in the team plays different position")
    for case in case_list:
        all_mixed = True
        for i in range(len(list(case))):
            if each_team_position[i] == list(case)[i]:
                all_mixed = False
        
        if all_mixed:
            print(list(case))
            each_team_case+=1

    if total_case == 0:
        total_case = each_team_case
    else:
        total_case*= each_team_case
        
    print("\n")
    print(team, "case", each_team_case)
    print("Total case : ",total_case)
    print("\n\n")
```
